```python
# splitting the dataser into dependent and independent variable
#independent variable as x
x=df.iloc[:,0:-1]
#dependent variable as y
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
#Train and Test the model 
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc =[]
model =[]
```

# Support vector machine

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(xtrain,ytrain)
#Evalution model
svmpred =svm.predict(xtest)
x=metrics.accuracy_score(ytest,svmpred)
acc.append(x)
model.append("SVM")
print("SVM Accuracy is : ",x*100)
print(classification_report(ytest,svmpred))
```

```python
svmcm = metrics.confusion_matrix(svmpred,ytest)
svmcm
```

```python
#saving trained SVM
import pickle
pkl_filename_SVM ='SVM.sav'
Model_pkl_SVM = open(pkl_filename_SVM,'wb')
pickle.dump(svm,Model_pkl_SVM)
Model_pkl_SVM.close()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier()
RF.fit(xtrain,ytrain)
#Evalution model
RFpred =RF.predict(xtest)
x=metrics.accuracy_score(ytest,RFpred)
acc.append(x)
model.append("RF")
print("RF Accuracy is : ",x*100)
print(classification_report(ytest,RFpred))
```

```python
RFcm = metrics.confusion_matrix(RFpred,ytest)
RFcm
```

```python
#saving trained RF
import pickle
pkl_filename_RF ='RF.sav'
Model_pkl_RF = open(pkl_filename_SVM,'wb')
pickle.dump(RF,Model_pkl_RF)
Model_pkl_RF.close()
```
